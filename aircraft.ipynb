```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('MASTER.txt.gz', compression='gzip', index_col='N-NUMBER', dtype={'COUNTRY': 'category', 'STATE': 'category', 'COUNTY': 'category'})
data
```

```python
data = data[data['COUNTRY'] == 'US']
data
```

```python
data = data[['COUNTRY', 'STATE', 'COUNTY']]
data
```

```python
county_data = data[['STATE', 'COUNTY']].apply(lambda x: ''.join(x), axis=1).astype('category')
```

```python
#data = df = pd.DataFrame(data[['STATE', 'COUNTY']])
#data['STATE_COUNTY'] = county_data
#data

out = pd.DataFrame(county_data.value_counts(), columns=['count'])
```

```python
data['STATE']
```

```python
counties = out.index.to_series().str.extract(r'([A-Z]+)(\d+)')
counties[1] = counties[1].str.pad(3, side='left', fillchar='0')

counties['count'] = out['count']
counties.columns = ['usps_state', 'fips_county', 'count']

counties = counties[counties['usps_state'].notnull()]

# compound key
state_county = counties[['usps_state', 'fips_county']].apply(lambda x: ''.join(x), axis=1).astype('category')
counties['state_county'] = state_county
counties

counties.to_csv('aircraft_county.csv')
```

```python
# now we load the populations
pop = pd.read_csv('population.csv', index_col=0, dtype={'fips_state': 'string', 'fips_county': 'string'})
pop
```

```python
# compound key
pop_state_county = pop[['usps_state', 'fips_county']].apply(lambda x: ''.join(x), axis=1).astype('category')
pop['state_county'] = pop_state_county
pop
```

```python
# join the two datasets
cols = pop[['county_name', 'population_2017']]
cols.index = pop['state_county']

out_data = counties.join(cols, on='state_county', how='inner')
out_data
```

```python
# now load the areas
areas = pd.read_csv('county_area.csv', index_col=0)
areas
```

```python
# and join them
cols = areas[['LND110210D']]
cols.index = areas['state_county']
cols
```

```python
new_out = out_data.join(cols, on='state_county', how='inner')
new_out = new_out.rename(columns={'LND110210D': 'county_area'})

new_out['population_density'] = new_out['population_2017'] / new_out['county_area']
new_out['aircraft_per_person'] = new_out['count'] / new_out['population_2017']
new_out
```

```python
new_out.to_csv('aircraft_analysis.csv')
```
