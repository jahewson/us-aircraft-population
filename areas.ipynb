```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('LND01.csv', dtype={'STCOU': 'string'})

# filter out entire country / states
data = data[data['Areaname'].str.contains(',')]

# 2010 census
data = data[['Areaname', 'STCOU', 'LND110210D']]
```

```python
data['usps_state'] = data['Areaname'].str.extract(r'([A-Z]{2})')
```

```python
data['fips_county'] = data['STCOU'].str.extract(r'(\d{2}(\d+))')[1]
data
```

```python
data['state_county'] = data[['usps_state', 'fips_county']].apply(lambda x: ''.join(x), axis=1).astype('category')
data
```

```python
data.to_csv('county_area.csv')
```
