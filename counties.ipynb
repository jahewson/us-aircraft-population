```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('co-est2017-alldata.csv', encoding='cp1252')
data = data[data['COUNTY'] != 0] # filter out states

# pluck columns
data = data[['STATE', 'COUNTY', 'CTYNAME', 'POPESTIMATE2017']]

# format 2-digit FIPS
data['STATE'] = data['STATE'].astype(str).str.pad(2, side='left', fillchar='0')

# format 3-digit FIPS
data['COUNTY'] = data['COUNTY'].astype(str).str.pad(3, side='left', fillchar='0')


data.columns = ['fips_state', 'fips_county', 'county_name', 'population_2017']
data
```
